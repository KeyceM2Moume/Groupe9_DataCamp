```python
import pandas as pd
from google_play_scraper import reviews, reviews_all, Sort
```

```python
# URL de l'application √† scraper (ici, Spotify)
url = 'com.spotify.music'

# Extraction des avis via l'API google-play-scraper
avis, _ = reviews(url, # Identifiant de l'application sur le Play Store
                  lang='fr', # Langue des avis (fran√ßais)
                  country='cm', # Pays des avis (Cameroun)
                  sort=Sort.MOST_RELEVANT, # Trier par pertinence (avis les plus pertinents)
                  count=10000, # Nombre maximal d'avis √† r√©cup√©rer
                  filter_score_with=None # Aucun filtre sur les notes (prendre tous les avis)
                 )
```

```python
avis
```

```python
# Conversion des avis r√©cup√©r√©s en DataFrame pandas
df = pd.DataFrame.from_records(avis)
df
# Sauvegarde des avis dans un fichier CSV
df.to_csv('scrapped_reviews.csv', index=False, encoding='utf-8-sig', sep=';')
```
